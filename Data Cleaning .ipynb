```python
null_var=df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_columns=null_var[null_var>20].keys()
drop_columns
```

```python
df2_drop_clm=df.drop(columns=drop_columns)
df2_drop_clm.shape
```

```python
df3_num=df2_drop_clm.select_dtypes(include=["int64","float64"])
df3_num.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
#to find the missing value where null value are present
missing_num_var=[var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
#to fill value where null valuees are present by using data distribution
plt.figure(figsize=(16,9))
sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':5,'color':'#dc143c'})
   
```

```python
df4_num_mean=df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':8,'color':'red'},label="original")
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':5,'color':'green'},label="mean"
plt.legend()
```

```python
df5_num_median=df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'red'}
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'green'}
    sns.distplot(df5_num_median[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'k'}
plt.legend()
```
